### Nearest Neighbour

```python
# Functions
def nn_resize(base_image, scale):
    print("Starting Nearest Neighbour")
    x_scale = int(scale*base_image.shape[0])
    y_scale = int(scale*base_image.shape[1])
    resized = np.zeros((x_scale,y_scale,3),  dtype=np.uint8)
    for x in range(0, x_scale):
        for y in range(0, y_scale):
            x_old = int(round(x/scale))
            if x_old >= base_image.shape[0]:
                x_old = base_image.shape[0] - 1
            y_old = int(round(y/scale))
            if y_old >= base_image.shape[1]:
                y_old = base_image.shape[1] - 1
            resized[x, y, :] =  base_image[x_old,y_old,:]
    print("Done resizing")
    return resized

def show_diff(original, resized, scale):
    plt.axis('off')
    print("Original 100 x 100")
    plt.imshow(original[:100, :100,:])
    plt.show()
    plt.axis('off')
    print("Resized by x{}, {} x {}".format(scale, int(scale*100), int(scale*100)))
    plt.imshow(resized[:int(scale*100), :int(scale*100),:])
    plt.show()
```

```python
# Enlarge
large_nn = nn_resize(cat, 3)
show_diff(cat, large_nn, 3)
plt.imshow(large_nn)
plt.show()
plt.imshow(cat)
plt.show()

# Shrink
small_nn = nn_resize(cat, 0.5)
show_diff(cat, small_nn, 0.5)
plt.imshow(small_nn)
plt.show()
plt.imshow(cat)
plt.show()
```

```python
# Enlarge
large_nn = nn_resize(jorah, 10)
print("Resized")
plt.imshow(large_nn)
plt.show()
print("Original")
plt.imshow(jorah)
plt.show()

# Shrink
large_nn = nn_resize(jorah, 0.1)
print("Resized")
plt.imshow(large_nn)
plt.show()
print("Original")
plt.imshow(jorah)
plt.show()
```

### Bilinear Interpolation

```python
# Functions
def bl_resize(base_image, scale):
    print("Starting Bilinear Interpolation")
    x_scale = int(scale * base_image.shape[0])
    y_scale = int(scale * base_image.shape[1])
    resized = np.zeros((x_scale, y_scale, 3), dtype=np.uint8)
    for x in range(0, x_scale):
        for y in range(0, y_scale):
            x_old = x / scale
            if x_old >= base_image.shape[0]:
                x_old = base_image.shape[0] - 1
            y_old = y / scale
            if y_old >= base_image.shape[1]:
                y_old = base_image.shape[1] - 1

            xf = int(floor(x_old))
            xc = int(ceil(x_old))
            yf = int(floor(y_old))
            yc = int(ceil(y_old))
            c_ff = base_image[xf, yf, :].astype(dtype=int)
            c_cf = base_image[xc, yc, :].astype(dtype=int)
            c_fc = base_image[xf, yc, :].astype(dtype=int)
            c_cc = base_image[xc, yc, :].astype(dtype=int)

            resized[x, y, :] = (
                (yc - y_old) * (((xc - x_old) * c_ff) + (x_old - xf) * (c_cf))
                + (y_old - yf) * (((xc - x_old) * c_fc) + ((x_old - xf) * c_cc))
            ).astype(dtype=np.uint8)

    print("Done resizing")
    return resized
```
